## Frequência de imóveis

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')

dados.head(10)
```

```python
# Seleção de imóveis classificados como Apartamento

selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao].shape[0]
n1
```

```python
# Seleção de imóveis classificados como 'Casa', 'Casa de Condomínio' e 'Casa de Vila'
# OR (I) / AND (&)

selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0]
n2
```

```python
# Seleção de imóveis com área entre 60 e 100 metros quadrados.

selecao = (dados['Area'] >=60) & (dados['Area'] <=100)
n3 = dados[selecao].shape[0]
n3
```

```python
# Seleção de imóveis com 4 quartos ou mais e aluguel menor que R$ 2.000,00

selecao = (dados['Quartos'] >=4) & (dados['Valor'] <=2000)
n4 = dados[selecao].shape[0]
n4
```

## Insight de frequência de imóveis

```python
print(f'Seleção de imóveis classificados como Apartamento: {n1}')
print(f'Seleção de imóveis classificados como Casa, Casa de Condomínio e Casa de Vila: {n2}')
print(f' Seleção de imóveis com área entre 60 e 100 metros quadrados: {n3}')
print(f'Seleção de imóveis com 4 quartos ou mais e aluguel menor que R$ 2.000,00: {n4}')
```

## Criando agrupamentos (Groupby)

```python
df = pd.read_csv('dados/aluguel.csv', sep=';')

df.head(10)
```

```python
df['Valor'].mean()
```

```python
bairros = df['Bairro'].unique()
selecao = df['Bairro'].isin(bairros)
df = df[selecao]
```

```python
df['Bairro'].drop_duplicates()
```

```python
grupo_bairro = df.groupby('Bairro')
```

```python
grupo_bairro.groups
```

```python
for bairro, dados, in grupo_bairro:
    print(f'Valor médio do bairro {bairro} -> {dados.Valor.mean()}')
```

```python
grupo_bairro[['Valor','Condominio']].mean().round(2)
```

## Tratamento de Outliers

```python
df = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
df.boxplot('Valor')
```

```python
df[df['Valor'] >= 500000]
```

```python
#Preparando Boxplot

valor = df['Valor']

Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
lim_inferior = Q1 - 1.5 * IIQ
lim_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >= lim_inferior) & (valor <= lim_superior)
df_new = df[selecao]

plt.rc('figure', figsize = (30,20))
df_new.boxplot(['Valor'])
```

```python
plt.rc('figure', figsize = (30,20))
df.hist(['Valor'])
df_new.hist(['Valor'])
```

```python
df.boxplot(['Valor'], by = ['Tipo'])
```
